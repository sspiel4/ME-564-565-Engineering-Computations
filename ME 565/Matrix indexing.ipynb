```python
L = 300
H = 300
c1=math.floor(L/2)
c2=math.floor(H/2)

rc1=math.floor(L/2-1)
rc2=math.floor(H/2-1)

matrix = np.zeros((H,L))

for i in range (0,c2):
    for j in range (0,c1):
        if (abs(i-(c2-1))**2+abs(j-(c1-1))**2)**0.5 > 0:
            matrix[i,j]=np.cos(abs(i-c2)/(abs(i-(c2-1))**2+abs(j-(c1-1))**2)**0.5)
        
for i in range (c2,H):
    for j in range (0,c1):
        if (abs(i-c2)**2+abs(j-(c1-1))**2)**0.5 > 0:
            matrix[i,j]=np.cos(abs(i-c2)/(abs(i-c2)**2+abs(j-(c1-1))**2)**0.5)
        
for i in range (0,c2):
    for j in range (c1,L):
        if (abs(i-(c2-1))**2+abs(j-c1)**2)**0.5 > 0:
            matrix[i,j]=np.cos(abs(i-c2)/(abs(i-(c2-1))**2+abs(j-c1)**2)**0.5)
        
for i in range (c2,H):
    for j in range (c1,L):
        if (abs(i-c2)**2+abs(j-c1)**2)**0.5  > 0:
            matrix[i,j]=np.cos(abs(i-c2)/(abs(i-c2)**2+abs(j-c1)**2)**0.5)
```

```python
fig=plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.imshow(matrix, extent=(0, L, 0, H),interpolation='nearest',cmap=cm.gist_rainbow)

plt.colorbar()
plt.show()
```

```python
L = 300
H = 300
c1=math.floor(L/2)
c2=math.floor(H/2)

rc1=math.floor(L/2-1)
rc2=math.floor(H/2-1)
r2=math.floor(rc1**2)

matrix = np.zeros((H,L))

for i in range (0,c2):
    for j in range (0,c1):
        if((i-c2)**2 +(j-c1)**2 >= r2):    # top left
            if (abs(i-(c2-1))**2+abs(j-(c1-1))**2)**0.5 > 0:
                matrix[i,j]=np.cos(abs(i-c2)/(abs(i-(c2-1))**2+abs(j-(c1-1))**2)**0.5)
        
for i in range (c2,H):
    for j in range (0,c1):
        if((i-c2)**2 +(j+1-c1)**2 >= r2):    # top right
            if (abs(i-c2)**2+abs(j-(c1-1))**2)**0.5 > 0:
                matrix[i,j]=np.cos(abs(i-c2)/(abs(i-c2)**2+abs(j-(c1-1))**2)**0.5)
        
for i in range (0,c2):
    for j in range (c1,L):
        if((i+1-c2)**2 +(j+1-c1)**2 >= r2):    # bottom right
            if (abs(i-(c2-1))**2+abs(j-c1)**2)**0.5 > 0:
                matrix[i,j]=np.cos(abs(i-c2)/(abs(i-(c2-1))**2+abs(j-c1)**2)**0.5)
        
for i in range (c2,H):
    for j in range (c1,L):
        if((i+1-c2)**2 +(j-c1)**2 >= r2):    # bottom left
            if (abs(i-c2)**2+abs(j-c1)**2)**0.5  > 0:
                matrix[i,j]=np.cos(abs(i-c2)/(abs(i-c2)**2+abs(j-c1)**2)**0.5)
```

```python
fig=plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.imshow(matrix, extent=(0, L, 0, H),interpolation='nearest',cmap=cm.gist_rainbow)

plt.colorbar()
plt.show()
```
