```python
import numpy as np

import control

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

### (a) (iii)

```python
#ODE: \ddot{x} + d*\dot{x} +k*x = u(t)

d = 5
k = 6

num = [1, 5]
den = [1, d, k]
sysG = control.tf(num, den)

#Calculate Impulse Response
[t,x] = control.impulse_response(sysG)

#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(t,x, label = 'Displacement')

#title axis
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Impulse Response')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

#Calculate Impulse Response
[t,x] = control.step_response(sysG)

#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(t,x, label = 'Displacement')

#title axis
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Step Response')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
