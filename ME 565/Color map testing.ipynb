```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.cm as cm
```

```python
color = np.random.rand(5,5)
color
```

```python
nrows, ncols = 5, 5
fig=plt.figure(figsize=(7, 7), dpi= 80, facecolor='w', edgecolor='k')
plt.imshow(color, extent=(0, 5, 0, 5),interpolation='nearest', cmap=cm.gist_rainbow)
plt.show()
```

```python
BC = np.zeros((100,100))
BC
```

```python
BC[0] = 1
BC[99] = 1
BC[:,0] = 1
BC[:,99] = 1
BC
```

```python
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.imshow(BC, extent=(0, 100, 0, 100),interpolation='nearest')
plt.colorbar()
plt.show()
```
