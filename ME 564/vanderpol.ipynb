```python
import numpy as np
from scipy import integrate
from numpy import linalg as LA

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
#Set up variables

'''
EQs
d/dt(y_1) = y_2
d/dt(y_2) = e*((y_1**2)-1)*y_2 - y_1

for inputs
y_1 = Y[0]
y_2 = Y[1]

d/dt(y_1) = Y[1]
d/dt(y_2) = e*((Y[0]**2)-1)*Y[1] - Y[0]
'''
```

```python
#Variables for part A

e=20

def solvr(Y, t):
    return [Y[1], -e*((Y[0]**2)-1)*Y[1] - Y[0]]
```

```python
#establish a time range to compute the solution
a_t = np.arange(0, 30, 0.01)
#matrix returned from solvr, initial conditions, time range.
asol = integrate.odeint(solvr, [0.1,1], a_t)
y_1 = asol[:,0:1]
y_2 = asol[:,1:2]
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(y_1,y_2, label = 'Position')

#title axis
ax.set_xlabel('y_1')
ax.set_ylabel('y_2')
ax.set_title('Final Exam 2c: Phase Portrait - Epsilon = {}'.format(e))

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/Final/plots/2c Epsilon={}.pdf'.format(e),
            bbox_inches = 'tight')
```
