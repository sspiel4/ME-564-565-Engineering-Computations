```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
#define j as imaginary number
j = (-1)**(.5)
```

```python
#create array of integers from 0 to 5 to use in plots. 
x = np.arange(0,10,.01)
```

```python
#Part A

#equation
y = np.exp(0)*((np.cos(x)))    #real part
y_i = np.exp(0)*((np.sin(x)))  #imaginary part
y_1 = np.exp(j*x)              #actual function


#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y_1, label = 'Original Function', lw = 1)
ax.plot(x,y, label = 'Analytical Solution', color = 'r', ls = '--', lw = 3)
ax.plot(x,y_i, label = 'Imaginary Solution', color = 'b', ls = '--', lw = 1)

#set limits on x axis
ax.set_xlim(0,10)
#ax.set_ylim(0,30)

#title axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Excercise 1-6 (a)')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

#show plot
plt.show()

#save PDF
fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW1/plots/1-6a.pdf',bbox_inches = 'tight')
```
