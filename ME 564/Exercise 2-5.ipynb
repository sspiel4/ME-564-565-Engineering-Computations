```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
#create array of integers from 0 to 5 to use in plots. 
x = np.arange(-5,5,.01)

#actual equation to check the plot
y_1 = 1/(-x+1)

#negative taylor series expansion
t_0 = 0
t_1 = -(1/x)
t_2 = -(x**-2)/2
t_3 = -(x**-3)/6
t_4 = -(x**-4)/24
t_5 = -(x**-5)/120
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y_1, label = 'Real Function')
ax.plot(x,t_0 + t_1, label = '1st Term')
#ax.plot(x,t_1 + t_2, label = '2 Terms')
#ax.plot(x,t_0 + t_1 + t_2 + t_3, label = '3 Terms')
#ax.plot(x,t_1 + t_2 + t_3 + t_4, label = '4 Terms')
#ax.plot(x, t_0 + t_1 + t_2 + t_3 + t_4 + t_5, label = '5 Terms')


#set limits on x axis
ax.set_xlim(-5,5)
ax.set_ylim(-4,4)

#title axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Excercise 2-5: Taylor Expansion')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

#save PDF
#fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW1/plots/1-3a.pdf',bbox_inches = 'tight')
```
