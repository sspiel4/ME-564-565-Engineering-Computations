```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create array of integers from 0 to 100 to use in plots. 
a = 0.001
b = 10.001
dx = 0.001
x = np.arange(a,b,dx)
f = np.sin(x)

'''
Note that some numpy does some strange things with 'f' here. You can look at f[-1] through f[-100]
and it is the same as f[0] through f[99]. So if you want to be 1 indexed on a vector, you can do that
by making it negative. 
'''

#find length of vector x
n = len(x)
```

```python
#for left rectangle 1.86
areaLR = 0
for i in range(0,n-1):
    areaLR = areaLR + dx*f[i]
print(areaLR)

#for right rectangle 1.81
areaRR = 0
for i in range(0,n):
    areaRR = areaRR + dx*f[i]
print(areaRR)
```

```python
#trapazoidal
i=0
areaTZ = 0
for i in range(1,n):
    areaTZ = areaTZ + (dx/2)*(f[i-1]+f[i])
areaTZ
```

```python
np.trapz(f,x,dx)
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])
ax.plot(x,f, label = 'f')

#set limits on x axis
ax.set_xlim(a,b)

#title axis
ax.set_xlabel('x')
ax.set_ylabel('f')
ax.set_title('Numerical Integration')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

#show plot
plt.show()
```
