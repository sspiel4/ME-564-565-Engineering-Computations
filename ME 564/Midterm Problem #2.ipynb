```python
import numpy as np
from scipy import integrate
from numpy import linalg as LA

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
def solvr(Y, t):
    return [-.01*Y[0], Y[0] + -0.011*Y[1]]
```

```python
A = np.array([[-.01,0],[1,-.011]])
```

```python
vals, vects = LA.eig(A)
print(vects)
print(vals)
```

```python
#establish a time range to compute the solution
a_t = np.arange(0, 500, 0.01)
#matrix returned from solvr, initial conditions, time range.
asol = integrate.odeint(solvr, [0.1, 0], a_t)
p_1 = asol[:,0:1]
v_1 = asol[:,1:2]
```
