```python
import numpy as np
from scipy import integrate
from numpy import linalg as LA

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
def solvr(Y, t):
    return [Y[1], Y[2], 6*Y[0] + 7*Y[1]]
```

```python
A = np.array([[0,1,0],[0,0,1],[6,7,0]])
print(A)
```

```python
eigs, e_vect = LA.eig(A)
print(eigs)
print(e_vect)
```

Solve and Plot ODE

```python
#establish a time range to compute the solution
a_t = np.arange(-1, 10, 0.01)
#matrix returned from solvr, initial conditions, time range.
asol = integrate.odeint(solvr, [1, -1, 1], a_t)
p_1 = asol[:,0:1]
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(a_t,p_1)

#title axis
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Excercise 2-2: Initial [1, -1, 1]')

plt.show()

fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW2/plots/2-2a.pdf',
            bbox_inches = 'tight')
```
