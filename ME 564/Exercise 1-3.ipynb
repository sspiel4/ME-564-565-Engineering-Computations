```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create array of integers from 0 to 5 to use in plots. 
x = np.arange(-5,5,.01)

#actual equation to check the plot
y_1 = (np.sin(x))/x

#taylor series expansion
t_1 = np.ones(1000)
t_2 = -(x**2)/6
t_3 = (x**4)/120
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y_1, label = 'Real Function')
ax.plot(x,t_1, label = '1st Term')
ax.plot(x,t_1 + t_2, label = '1st, 2nd Terms')
ax.plot(x,t_1 + t_2 + t_3, label = '1st, 2nd, 3rd Terms')

#set limits on x axis
ax.set_xlim(-5,5)
ax.set_ylim(-1,2)

#title axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Excercise 1-3 (a): Taylor Expansion [sin(x)/x]')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

#save PDF
fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW1/plots/1-3a.pdf',bbox_inches = 'tight')
```

```python
#create array of integers from 0 to 5 to use in plots. 
x = np.arange(-5,5,.01)

#actual equation to check the plot
y2_1 = 3**x

#taylor series expansion
t2_1 = np.ones(1000)
t2_2 = x*np.log(3)
t2_3 = ((x**2)*np.log(3**2))/2
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y2_1, label = 'Real Function')
ax.plot(x,t2_1, label = '1st Term')
ax.plot(x,t2_1 + t2_2, label = '1st, 2nd Terms')
ax.plot(x,t2_1 + t2_2 + t2_3, label = '1st, 2nd, 3rd Terms')

#set limits on x axis
ax.set_xlim(-5,5)
#ax.set_ylim(-1,5)

#title axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Excercise 1-3 (b): Taylor Expansion [3^x]')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

#save PDF
fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW1/plots/1-3b.pdf',bbox_inches = 'tight')
```
