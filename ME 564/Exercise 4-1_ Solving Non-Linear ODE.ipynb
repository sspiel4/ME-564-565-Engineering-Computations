### Part A - No Forcing

```python
#Variables for part A

a=1
b=-1
d=0.2
g=0
w=0

def solvr(Y, t):
    return [Y[1], -d*Y[1]-b*Y[0]-a*Y[0]**3+g*np.cos(w*t)]
```

```python
#establish a time range to compute the solution
a_t = np.arange(0, 100, 0.01)
#matrix returned from solvr, initial conditions, time range.
asol = integrate.odeint(solvr, [2,0], a_t)
p_1 = asol[:,0:1]
v_1 = asol[:,1:2]
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(a_t,p_1, label = 'Position')
ax.plot(a_t,v_1, label = 'Velocity')

#title axis
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Excercise 4-1: Part A - No Forcing')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW4/plots/4-1a.pdf',
            bbox_inches = 'tight')
```
