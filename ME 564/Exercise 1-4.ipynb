```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
#create array of integers from 0 to 5 to use in plots. 
x = np.arange(-100,100,.1)
j = (-1)**0.5

#actual equation to check the plot
y_1 = np.exp(-.2*x) + np.exp(.2*x)
y_2 = np.exp(2*x) + np.exp(-2*x)
y_3 = np.exp(-j*.2*x) + np.exp(.2*j*x)
y_4 = np.exp(2*j*x) + np.exp(2*j*x)
y_5 = np.exp(-2*x)

```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y_1, label = 'y_1')
ax.plot(x,y_2, label = 'y_2')
ax.plot(x,y_3, label = 'y_3')
ax.plot(x,y_4, label = 'y_4')
ax.plot(x,y_5, label = 'y_5')

#set limits on x axis
ax.set_xlim(-50,50)
ax.set_ylim(-2.5,5)

#title axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Excercise 1-4: f(x)=c_1*e^(at)+c_2*e^(-at)')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

#save PDF
fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW1/plots/1-4.pdf',bbox_inches = 'tight')
```
