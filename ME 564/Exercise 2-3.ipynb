```python
import numpy as np
from scipy import integrate
from numpy import linalg as LA

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

## Coupled 1st Order System in Matrix Form

```python
A = np.array([[0,1,0,0],[-2,0,1,0],[0,0,0,1],[1,0,-1,0]])
vals, vects = LA.eig(A)
```

# Eigen Values

```python
vals
```

```python
vects
```

## Plot EQ

```python
def solvr(Y, t):
    return [Y[1], -2*Y[0] + Y[2], Y[3], Y[0] - 1*Y[2]]
```

```python
#establish a time range to compute the solution
a_t = np.arange(0, 15, 0.01)
#matrix returned from solvr, initial conditions, time range.
asol = integrate.odeint(solvr, [1, 0, 1, 0], a_t)
p_1 = asol[:,0:1]
a_1 = asol[:,1:2]
p_2 = asol[:,2:3]
a_2 = asol[:,3:4]
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])

ax.plot(a_t,p_1, label = 'Weight 1')
ax.plot(a_t,p_2, label = 'Weight 2')

#title axis
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Excercise 2-3: Part C Solution - Position and Time')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW2/plots/2-3a.pdf',
            bbox_inches = 'tight')
```
