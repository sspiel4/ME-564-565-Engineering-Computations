```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
Plot cos(πmx) on the interval [0, 1] with dx = 0.01 for m = 0, 1, 2, 3, 4, and 5. Verify
numerically, using trapezoidal integration (i.e. trapz), that cos(πmx) and cos(πnx) are
orthogonal for the following (m, n) pairs: (1, 4), (2, 6), and (3, 15).
'''

dx = 0.01
a = 0
b = 1 + dx
x = np.arange(a,b,dx)
```

```python
i = 0
for i in range(0,5):
    f = np.cos(np.pi*x*i)
    
    #create figure object
    fig = plt.figure()

    #plot each set of y values with respect to time
    ax = fig.add_axes([0,0,1,1])

    ax.plot(x,f, label = i)

    #title axis
    ax.set_xlabel('x')
    ax.set_ylabel('f')
    ax.set_title('Excercise 5-2(c) where m={}'.format(i))

    #include legend
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    plt.show()
    
    #save PDF
    fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW5/plots/5-2c m={}.pdf'.format(i),
        bbox_inches = 'tight')
    
    i = i + 1
```

```python
#show that <f,g> are orthogonal for pairs: (1, 4), (2, 6), and (3, 15)

#define variables m, and n
m = 1
n = 4

f = np.cos(np.pi*x*m)
g = np.cos(np.pi*x*n)

fg = np.cos(np.pi*x*m) * np.cos(np.pi*x*n)
np.trapz(fg, x, dx)
```
