```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
#create array of integers from 0 to 100 to use in plots. 
t = np.arange(0,5,.01)

#set variables for lambda coefficients
lambda_1=-5
lambda_2=-1
lambda_3=0
lambda_4=0.01
lambda_5=0.1

#calculate y values for each array of lambda values
y_1 = np.exp(lambda_1*t)*2
y_2 = np.exp(lambda_2*t)*2
y_3 = np.exp(lambda_3*t)*2
y_4 = np.exp(lambda_4*t)*2
y_5 = np.exp(lambda_5*t)*2

#note that the +2 should really be outside the parenthesis.
#however, putting the +2 outside the parenthesis causes
# a negative value in the ln() which results in all values as 
#NAN. how is this avoided?
```

```python
#create figure object
fig = plt.figure()

#plot each set of y values with respect to time
ax = fig.add_axes([0,0,1,1])
ax.plot(t,y_1, label = 'lambda 1')
ax.plot(t,y_2, label = 'lambda 2')
ax.plot(t,y_3, label = 'lambda 3')
ax.plot(t,y_4, label = 'lambda 4')
ax.plot(t,y_5, label = 'lambda 5')

#set limits on x axis
ax.set_xlim(0,5)

#title axis
ax.set_xlabel('time (t)')
ax.set_ylabel('mass (m)')
ax.set_title('Excercise 1-2: Half Life of a Radioactive Element')

#include legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

#show plot
plt.show()

#save PDF
fig.savefig('C:/Users/sspie/OneDrive/Documents/1_Graduate School/ME 564 Eng Computation/HW1/plots/1-2.pdf',
            bbox_inches = 'tight')
```
